```python
import json
import urllib.request
import urllib.error
```

```python
def count_languages(data, d):
    for i in data:
        lang = i['language']
        if lang in d:
            d[lang] += 1
        else:
            d[lang] = 1
    return d
```

```python
def about_user(userlist, token):
    user = input('Введите имя пользователя: ')
    while user not in userlist:
        user = input('Пользователя нет в списке. Попробуйте ещё раз: ')
    print('Вы выбрали пользователя {}.\n'.format(user))
    url = 'https://api.github.com/users/%s/repos?per_page=100&access_token=%s' % (user, token) 
    try:
        response = urllib.request.urlopen(url)
    except urllib.error.HTTPError:
        print('Пользователя не существует. Введите список существующих пользователей.')
    except urllib.error.URLError:
        print('Отсутствует подключение к сети Интернет. Проверьте подключение и перезапустите программу.')
    else:
        text = response.read().decode('utf-8')
        data = json.loads(text)
        print('Его репозитории:')
        for i in data:
            if i['description'] == None:
                print('{}: без описания'.format(i['name']))
            else:
                print('{}: {}'.format(i['name'], i['description']))
        d = {}
        d = count_languages(data, d)
        print('\nПользователь пишет:')
        for k in d.keys():
            n = d[k]
            if k == None:
                if n == 1 or (n % 100 > 20 and n % 10 == 1):
                    print('\tбез использования языка программирования в {} репозитории'.format(n))
                else:
                    print('\tбез использования языка программирования в {} репозиторях'.format(n))
            elif n == 1 or (n % 100 > 20 and n % 10 == 1):
                print('\tна языке {} в {} репозитории'.format(k, n))
            else:
                print('\tна языке {} в {} репозиториях'.format(k, n))
```

```python
def count_repos(userlist, token):
    d = {}
    for user in userlist:
        url = 'https://api.github.com/users/%s/repos?per_page=100&access_token=%s' % (user, token) 
        try:
            response = urllib.request.urlopen(url)
        except urllib.error.HTTPError:
            d[user] = 0
        except urllib.error.URLError:
            print('Отсутствует подключение к сети Интернет. Проверьте подключение и перезапустите программу.')
        else:
            text = response.read().decode('utf-8')
            data = json.loads(text)
            d[user] = len(data)
    return d
```

```python
def find_max(d):
    maxn = 0
    maxu = []
    for k in d.keys():
        if d[k] >= maxn:
            maxn = d[k]
    for k in d.keys():
        if d[k] == maxn:
            maxu.append(k)
    if len(maxu) == 1:
        print('\nБольше всего репозиториев ({}) у пользователя {}.'.format(maxn, maxu[0]))
    else:
        print('\nБольше всего репозиториев ({}) у пользователей {}.'.format(maxn, ', '.join(maxu)))
```

```python
def best_lang(userlist, token):
    d = {}
    for user in userlist:
        url = 'https://api.github.com/users/%s/repos?per_page=100&access_token=%s' % (user, token) 
        try:
            response = urllib.request.urlopen(url)
        except urllib.error.HTTPError:
            pass
        except urllib.error.URLError:
            print('Отсутствует подключение к сети Интернет. Проверьте подключение и перезапустите программу.')
        else:
            text = response.read().decode('utf-8')
            data = json.loads(text)
            count_languages(data, d)
    maxn = 0
    bestl = []
    for k in d.keys():
        if d[k] >= maxn:
            maxn = d[k]
    for k in d.keys():
        if d[k] == maxn:
            bestl.append(k)
    if len(bestl) == 1:
        print('\nСамый популярный язык среди пользователей: {}.'.format(bestl[0]))
    else:
        print('\nСамые популярные языки среди пользователей: {}.'.format(', '.join(bestl)))
```

```python
def followers(userlist, token):
    d = {}
    for user in userlist:
        url = 'https://api.github.com/users/%s/followers?per_page=100&access_token=%s' % (user, token)
        try:
            response = urllib.request.urlopen(url)
        except urllib.error.HTTPError:
            pass
        except urllib.error.URLError:
            print('Отсутствует подключение к сети Интернет. Проверьте подключение и перезапустите программу.')
        else:
            text = response.read().decode('utf-8')
            data = json.loads(text)
            d[user] = len(data)
    maxn = 0
    maxf = []
    for k in d.keys():
        if d[k] >= maxn:
            maxn = d[k]
    for k in d.keys():
        if d[k] == maxn:
            maxf.append(k)
    if len(maxf) == 1:
        print('\nБольше всего подписчиков ({}) у пользователя {}.'.format(maxn, maxf[0]))
    else:
        print('\nБольше всего подписчиков ({}) у пользователей {}.'.format(maxn, ', '.join(maxf)))
```

```python
def ask_for_users():
    users = input('Введите список пользователей через запятую без пробелов: ').split(',')
    return users
```

```python
def main(userlist, token):
    print('Пользователи:' + ', '.join(userlist))
    about_user(userlist, token)
    find_max(count_repos(userlist, token))
    best_lang(userlist, token)
    followers(userlist, token)
```
